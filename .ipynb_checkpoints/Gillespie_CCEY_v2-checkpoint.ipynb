## Defining total molecule counts as global constants

```python
#---------------------------------# 
# Total molecular counts -#-

R_tot = 11400       # Total number of RNAP molecules
pLac_tot = 7        # Total number of plasmids in e-coli
pTet_tot = pLac_tot # Since both TetR and LacI are on the same plasmid
LacI_tot = 6        # Total number of LacI Repressor molecules
aTc_tot = 135000    # Equivalent to 100 ng/uL of aTc
TetR_tot = 30       # Total number of TetR repressor molecules
#---------------------------------# 
# mSpinach = mRNA for CFP
# mG aptamer = mRNA for RFP
# Length scales

TLs = 818    # Transcript length for CFP
PLs = 40     # Promoter length for CFP
TLg = 758    # Transcript length for RFP
PLg = 44     # Promoter length for RFP
h0 = 10.5    # Number of bases of DNA/twist for B-form DNA
Ns = 150     # Intergenic spacer length    
pLen = 2500  # Plasmid length for global supercoiling state

#---------------------------------# 
# Maintenance term components

T0 = 1.2     # Number of topoisomerase molecules
G0 = 7.23    # Number of Gyrase molecules
tau = 0.5    # Negative coils introduced per second/ Topo   
gamma = 0.5  # Positive coils introduced per second/ Topo
K_Mtau = 120    # Hill constant for topoisomerase activity
K_Mgamma = 120  # Hill constant for gyrase activity
sigma0 = -0.065 # Base level of supercoiling density on genome
rhol = 0.1167
rhoT = 0.1167   # Constitutive production rates/min
#---------------------------------# 
# Translation terms

kB16 = 0.75     # Ribosomal affinity for mCFP
k_fC = 0.000556 # CFP folding rate
dProtein = 0.0001667
k_fR = 0.0001515# RFP folding rate
k_f1R = 0.001667# First folding rate of RFP
f_f2R = 0.003334# Second folding rate of RFP
k_BgL = 1.25    # Ribosomal affinity for mRFP
k_TL0 = 0.023   # Averaged translation rate of CFP and RFP
Ribo_tot = 6800 #Total number of Ribosome machines
alpha0 = 0.000627

#---------------------------------# 
# Rate constants (1/s)

dm = 0.0001672        # mCFP degradation rate
k_open = 0.04         # Open complex formation rate
k_r    = 550          # Reverse transcription rate
k_uaT  = 1.           # aTc-TetR Dissociation rate
k_uT   = 10.          # Fall off rate of TetR
k_uL   = 0.022        # Fall off rate of LacI
k_aL   = 6000         # IPTG bind rate
k_seqLac = 10         # LacI binding rate
k_seqTet = 10         # TetR binding rate
dp     = 8.362e-5     # mRFP degradation rate
k_aT   = 6000         # aTc bind rate
```

## Calculating Propensities

```python
def simple_propensity(population, t):
    """
    Returns an array of propensities given a set of parameters
    and an array of populations.
    """
    # Unpack parameters
    # alpha1, beta, gamma, eta, K, iptg = params
    # basal_rfp, rep_tot, rep_coop, Kd, aTc_coop, rfp_degrade, basal_cfp, cfp_degrade, aTc, super_prop, super_density = params
    #dm, k_open, k_r, k_uaT, k_uT, k_uL, k_aL, k_seqLac, k_seqTet, dp, k_aT = params

    # Unpack population
    mS, mG, ECs, ECg, CCs, CCg, LacI, TetR, aTc, sigma_ts, sigma_tg, sigma_ps, sigma_pg, P_cu, P_cf, P_ru, P_rf, Pr = population
    # Calculating k_cat(sigma) Rate constants in the flourescence equation
    #---------------------------------------------------------------------------------------------------# 
    k_catts = k_cat_sigma(sigma_ts, 0, 'ts')
    k_cattg = k_cat_sigma(sigma_tg, 0, 'tg')
    k_catps = k_cat_sigma(sigma_ps, 0, 'ps')
    k_catpg = k_cat_sigma(sigma_pg, 0, 'pg')
    #---------------------------------------------------------------------------------------------------# 
    # Calculating k_f(sigma) Rate constants in the closed complex equation
    k_fs = k_cat_sigma(sigma_ps, 1, 'pg')  # Third entry does not really matter here
    k_fg = k_cat_sigma(sigma_pg, 1, 'pg')
    #---------------------------------------------------------------------------------------------------# 
    # Calculating intermediate sums used in propensities
    R = R_tot - ECs - ECg - CCs - CCg
    pLac = pLac_tot + LacI - (LacI_tot + CCs + ECs)
    pTet = pTet_tot + TetR + aTc_tot - (CCg + ECg + aTc + TetR_tot)
    pTetC = TetR_tot - TetR - (aTc_tot - aTc)
    #---------------------------------------------------------------------------------------------------# 
    # Calculating lengths for supercoiling density rate equations
    n_fs = max(PLs + TLs + Ns + TLg*(TetR_tot - TetR - (aTc_tot - aTc))/(pTet_tot - CCg - ECg) - (sigma_ts + sigma_tg)/h0,0)
    #n_fs = 1
    n_fg = max(PLg + TLg + Ns + TLs*(LacI_tot - LacI)/(pLac_tot - CCs - ECs) + (sigma_ts + sigma_tg)/h0,0)
    #n_fg = 1
    alphaS = 0.5*TLs/(h0*n_fs)
    betaS = 0.5*PLs/(h0*n_fs)
    alphaG = 0.5*TLg/(h0*n_fg)
    betaG = 0.5*PLg/(h0*n_fg)
    
    #---------------------------------------------------------------------------------------------------# 
    # Calculating propensitie for terms in the supercoiling density rate equations
    
    fwd_sigma_tg = betaG*(k_cattg*ECg + k_fg*R*pTet) + alphaG*(2*dm*mG + k_open*CCg) + maint_calc(sigma_tg,1)
    fwd_sigma_ts = betaS*(k_catts*ECg + k_fs*R*pLac) + alphaS*(2*dm*mS + k_open*CCs) + maint_calc(sigma_ts,1)
    bwd_sigma_tg = betaG*(k_open*CCg + (k_r + k_open)*CCg) + 2*alphaG*k_cattg*ECg + maint_calc(sigma_tg,0)
    bwd_sigma_ts = betaS*(k_open*CCs + (k_r + k_open)*CCs) + 2*alphaS*k_catts*ECs + maint_calc(sigma_tg,0)
    fwd_sigma_pg = alphaG*(k_cattg*ECg + k_fg*R*pTet) + maint_calc(sigma_pg,1)
    bwd_sigma_pg = alphaG*(2*k_open + k_r)*CCg + maint_calc(sigma_pg,0)
    fwd_sigma_ps = betaS*(k_cattg*ECg + k_fg*R*pTet) + maint_calc(sigma_pg,1)
    bwd_sigma_ps = betaS*(2*k_open + k_r)*CCs + maint_calc(sigma_ps,0)
    
    #---------------------------------------------------------------------------------------------------# 
    # Translation propensities
    
    k_TL = k_TL0*np.exp(-alpha0*t)
    
    #---------------------------------------------------------------------------------------------------# 
    # All propensities (26)
    return np.array([k_catts*ECs,
                    dm*mS,
                    k_cattg*ECg,
                    dm*mG,
                    k_open*CCs,
                    k_catts*ECs,
                    k_open*CCg,
                    k_cattg*ECg,
                    k_fs*R*pLac,
                    (k_r + k_open)*CCs,
                    k_fg*R*pTet,
                    (k_r + k_open)*CCg,
                    rhoL + k_uL*(LacI_tot - LacI), 
                    (k_seqLac + dp)*LacI,
                    rhoT +  k_uaT*(aTc_tot - aTc) + k_uT*pTetC,
                    k_aL*TetR*aTc + (k_seqTet + dp)*TetR,
                    k_uaT*(TetR_tot - TetR - pTetC),
                    k_aT*(TetR + pTetC)*aTc,
                    fwd_sigma_ts,
                    bwd_sigma_ts,
                    fwd_sigma_tg,
                    bwd_sigma_tg,
                    fwd_sigma_ps,
                    bwd_sigma_ps,
                    fwd_sigma_pg,
                    bwd_sigma_pg,
                    kB16*k_fC*k_TL*Ribo_tot*mS,
                    k_fC*P_cu + dProtein*P_cu,
                    k_fC*P_cu,
                    k_BgL*k_fR*k_TL*Ribo_tot*mG,
                    k_fR*P_ru + dProtein*P_ru,
                    k_f1R*P_ru,
                    k_f2R*P_rf,
                    k_f2R*P_rf
                    ]), n_fs
```

## k cat/f calculator

```python
def k_cat_sigma(sigma, choice, event):
    beta = 0.02
    K_Msigma = 30.1
    if event == 'ts':
        zeta = 30.86   # Base transcription rate Del Vecchio
        sigma_star = sigma0*pLen/TLs
    elif event == 'tg':
        zeta = 28.60  # Base transcription rate Del Vecchio
        sigma_star = sigma0*pLen/TLg
    elif event == 'ps':
        zeta = 30.86   # Base transcription rate Del Vecchio
        sigma_star = sigma0*pLen/PLs
    else:
        zeta = 28.60   # Base transcription rate Del Vecchio
        sigma_star = sigma0*pLen/PLg
    if choice:
        return zeta/(1 + (sigma - sigma_star)**2/K_Msigma)
    else:
        return beta/(1 + (sigma - sigma_star)**2/K_Msigma)
```

## Maintenance term calculator

```python
def maint_calc(sigma, choice):
    if choice:
        return T0*tau*(sigma - sigma0)/(K_Mtau*sigma0 + (sigma - sigma0)**2)
    else:
        return G0*gamma*(sigma - sigma0)/(K_Mgamma*sigma0 + (sigma - sigma0)**2)
        
```

## Drawing the Reaction

```python
def sample_discrete(probs):
    """
    Randomly sample an index with probability given by probs.
    """
    # Generate random number
    q = np.random.rand()
    
    # Find index
    i = 0
    p_sum = 0.0
    while p_sum < q:
        p_sum += probs[i]
        i += 1
    return i - 1
```

## Returning draw and time

```python
# Function to draw time interval and choice of reaction
def gillespie_draw(params, propensity_func, population, t):
    """
    Draws a reaction and the time it took to do that reaction.
    """
    # Compute propensities
    props, n_fs_total = propensity_func(params, population, t)

    # Sum of propensities
    props_sum = props.sum()
    # Compute time

    time = np.random.exponential(1.0 / props_sum)
    
    # Compute discrete probabilities of each reaction
    rxn_probs = props / props_sum

    # Draw reaction from this distribution
    rxn = sample_discrete(rxn_probs)
    print("Reaction = ",rxn)
    return rxn, time, n_fs_total
```

## Main code

```python
def gillespie_ssa(params, propensity_func, update, population_0, 
                  time_points):
    """
    Uses the Gillespie stochastic simulation algorithm to sample
    from proability distribution of particle counts over time.
    
    Parameters
    ----------
    params : arbitrary
        The set of parameters to be passed to propensity_func.
    propensity_func : function
        Function of the form f(params, population) that takes the current
        population of particle counts and return an array of propensities
        for each reaction.
    update : ndarray, shape (num_reactions, num_chemical_species)
        Entry i, j gives the change in particle counts of species j
        for chemical reaction i.
    population_0 : array_like, shape (num_chemical_species)
        Array of initial populations of all chemical species.
    time_points : array_like, shape (num_time_points,)
        Array of points in time for which to sample the probability
        distribution.
        
    Returns
    -------
    sample : ndarray, shape (num_time_points, num_chemical_species)
        Entry i, j is the count of chemical species j at time
        time_points[i].
    """

    # Initialize output
    pop_out = np.empty((len(time_points), update.shape[1]), dtype=np.int)
    n_fs_total = np.empty((len(time_points)), dtype=np.int)
    # Initialize and perform simulation
    i_time = 1
    i = 0
    t = time_points[0]
    population = population_0.copy()
    pop_out[0,:] = population
    while i < len(time_points):
        #print(i)
        while t < time_points[i_time]:
            event, dt, n_fs_temp = gillespie_draw(params, propensity_func, population, t)
            #print(dt)
            # Update the population
            population_previous = population.copy()
            population += update[event,:]
            t += dt
            print(t)
        # Update the index
        n_fs_total[i] = n_fs_temp
        i = np.searchsorted(time_points > t, True)
        print(t)
        # Update the population
        pop_out[i_time:min(i,len(time_points))] = population_previous
        # Increment index
        i_time = i 
    return pop_out, n_fs_total
```
